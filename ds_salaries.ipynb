```python
# Salary Analysis 
# Overall Salary Stats

df['salary_in_usd'].describe()
```

```python
# Salary by Experience Level
# How does salary grow with experience?

df.groupby('experience_level')['salary_in_usd'].mean().sort_values()
```

```python
df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
```

```python
# Salary by Job Title (Top Paying Roles)
# What are the top-paying data science jobs? 

df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)
```

```python
# Remote Work Analysis
# Average Salary by Remote Ratio
# Does Remote work affect salary? 
# Remote ratio meanings: 
''' 
0 -> Onsite
50 -> Hybrid
100 -> Fully remote
'''

df.groupby('remote_ratio')['salary_in_usd'].mean()
```

```python
# Location-Based Analysis
# Salary by Company Location
# Which countries pay best? 

df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False)
```

```python
# Company Size Impact
# Do big companies pay more?
# Typical sizes:
'''
S - Small
M - Medium
L - Large
'''

df.groupby('company_size')['salary_in_usd'].mean()
```

```python
# Creating useful columns
# Salary banding 

df['salary_band'] = pd.cut(
    df['salary_in_usd'],
    bins=[0, 50000, 100000, 150000, 300000],
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

```python
df['salary_band'].value_counts()
```

```python
# Filtration
# Senior Remote Roles Only

df[
    (df['experience_level'] == 'SE') & (df['remote_ratio'] == 100)
]
```

```python
# Sorting and Presenting Data

df.sort_values(by='salary_in_usd', ascending=False).head(20)
```

```python
# Exporting Cleaned and Analyzed Dataset

df.to_csv('cleaned_ds_salaries.csv', index=False)
```
